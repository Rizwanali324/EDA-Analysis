# Cleanign and preprocessing


```python
df.drop(['fnlwgt','education','race'],axis=1,inplace=True)
df.sample(9)
```

```python
# Replace unknown values with mode for string columns
string_columns = df.select_dtypes(include='object').columns
for column in string_columns:
    mode_value = df[column].mode().iloc[0]
    df[column] = df[column].replace(r'\?', mode_value, regex=True)

# Replace unknown values with mean for integer columns
integer_columns = df.select_dtypes(include='number').columns
for column in integer_columns:
    mean_value = df[column].mean()
    df[column] = df[column].fillna(mean_value)
```

```python


# Set Seaborn style
sns.set(style="whitegrid")

# 1. Income Disparity Analysis
income_distribution = df['income'].value_counts()
sns.countplot(data=df, x='income')
plt.title("Income Distribution")
plt.show()
```

```python


# 2. Education and Income
edu_income_relation = df.groupby('educational-num')['income'].value_counts().unstack()
edu_income_relation.plot(kind='bar', stacked=True)
plt.title("Education vs. Income")
plt.xlabel("Educational Number")
plt.ylabel("Count")
plt.show()

```

```python
# 3. Occupation Insights
common_occupations = df['occupation'].value_counts()
sns.barplot(x=common_occupations.index, y=common_occupations.values)
plt.xticks(rotation=45)
plt.title("Most Common Occupations")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.show()

```

```python
# 4. Age and Marital Status
age_marital_distribution = df.groupby('marital-status')['age'].mean()
age_marital_distribution.plot(kind='bar')
plt.title("Average Age by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Average Age")
plt.show()

```

```python
# 5. Workclass and Hours Worked
workclass_hours = df.groupby('workclass')['hours-per-week'].mean()
sns.barplot(x=workclass_hours.index, y=workclass_hours.values)
plt.xticks(rotation=45)
plt.title("Average Hours Worked per Week by Workclass")
plt.xlabel("Workclass")
plt.ylabel("Average Hours Worked")
plt.show()
```

```python

# 6. Geographical Analysis
country_distribution = df['native-country'].value_counts().head(10)
country_distribution.plot(kind='bar')
plt.title("Top 10 Native Countries")
plt.xlabel("Native Country")
plt.ylabel("Count")
plt.show()
```

```python

# 7. Gender and Income Disparity
gender_income_gap = df.groupby('gender')['income'].value_counts()
gender_income_gap.unstack().plot(kind='bar', stacked=True)
plt.title("Gender vs. Income")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python


# 8. Capital Analysis
capital_distribution = df[['capital-gain', 'capital-loss']].describe()
capital_distribution.plot(kind='bar')
plt.title("Capital Gain and Loss Summary")
plt.ylabel("Value")
plt.show()
```

```python

# 9. Relationship Dynamics
relationship_income_hours = df.groupby(['relationship', 'income'])['hours-per-week'].mean().unstack()
relationship_income_hours.plot(kind='bar')
plt.title("Average Hours Worked per Week by Relationship and Income")
plt.xlabel("Relationship")
plt.ylabel("Average Hours Worked")
plt.xticks(rotation=45)
plt.show()
```

```python

# 10. Age and Work Patterns
age_hours_relation = df.groupby('age')['hours-per-week'].mean()
age_income_hours = df.groupby(['age', 'income'])['hours-per-week'].mean().unstack()
plt.scatter(age_hours_relation.index, age_hours_relation.values)
plt.plot(age_income_hours.index, age_income_hours['<=50K'], color='blue', label='Income <= 50K')
plt.plot(age_income_hours.index, age_income_hours['>50K'], color='orange', label='Income > 50K')
plt.title("Age vs. Average Hours Worked per Week")
plt.xlabel("Age")
plt.ylabel("Average Hours Worked")
plt.legend()
plt.show()
```

```python
# 1. Marital Status and Capital Strategies
marital_capital = df.groupby('marital-status')[['capital-gain', 'capital-loss']].mean()
marital_capital.plot(kind='bar')
plt.title("Marital Status and Average Capital Gain/Loss")
plt.xlabel("Marital Status")
plt.ylabel("Average Capital")
plt.show()
```

```python
# 3. Native Country and Career Trajectories
plt.figure(figsize=(50,50))
country_income_education = df.groupby(['native-country', 'income'])['educational-num'].mean().unstack()
country_income_education.plot(kind='bar', stacked=True)
plt.title("Native Country, Income, and Average Education Level")
plt.xlabel("Native Country")
plt.ylabel("Average Educational Number")
plt.xticks(rotation=45)
plt.show()

```
