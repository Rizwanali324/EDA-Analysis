# EDA question

```python
# 1. Distribution Analysis
sns.displot(data=df, x='Length', hue='Sex', multiple='stack')
plt.title("Distribution of Length across Genders")
plt.show()
```

```python
# 2. Correlation Exploration
correlation_matrix = df[['Whole weight', 'Shucked weight']].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation between Whole and Shucked Weight")
plt.show()
```

```python
# 4. Age Estimation
sns.scatterplot(data=df, x='Rings', y='Length', hue='Sex')
plt.title("Age Estimation based on Rings and Length")
plt.show()
```

```python
# 5. Weight Proportions
df['Weight Ratio'] = df['Shucked weight'] / df['Whole weight']
sns.boxplot(data=df, x='Sex', y='Weight Ratio')
plt.title("Weight Ratio by Gender")
plt.show()
```

```python
# 6. Dimensionality Reduction (PCA)
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df.drop('Sex', axis=1))

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

plt.plot(cumulative_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA: Cumulative Explained Variance')
plt.show()
```

```python
# 7. Gender Differences
sns.pairplot(df, hue='Sex', vars=['Length', 'Diameter', 'Height'])
plt.title("Gender Differences in Physical Attributes")
plt.show()
```

```python
# 8. Shell Characteristics
sns.scatterplot(data=df, x='Shell weight', y='Viscera weight',hue="Sex",size="Diameter")
plt.title("Shell Weight vs. Viscera Weight")
plt.show()
```

```python
# 9. Outlier Detection
sns.boxplot(data=df.drop('Sex', axis=1))
plt.title("Boxplot to Detect Outliers")
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Feature Importance (Random Forest)
from sklearn.ensemble import RandomForestRegressor

X = df.drop(['Rings', 'Sex'], axis=1)
y = df['Rings']

model = RandomForestRegressor()
model.fit(X, y)

importance = model.feature_importances_
plt.bar(X.columns, importance)
plt.xticks(rotation=45)
plt.title("Feature Importance using Random Forest")
plt.show()
```
